```python
%load_ext autoreload
%autoreload 2

import sys
sys.path.append("/home/jdli/")

import torch
from torch.utils.data import DataLoader
from transpec.model import Spec2label
from transpec.data import GaiaXPlabel_cont_v0
from matplotlib.colors import LogNorm
import cmasher as cmr
```

```python
data_dir = "/data/jdli/gaia/"
tr_file = "ap17_xpcont_cut.npy"

device = torch.device('cuda:1')


gdata  = GaiaXPlabel_cont_v0(data_dir+tr_file, part_train=False, device=device)
gdata_loader = DataLoader(gdata, batch_size=1024)
```

```python
model_tefflogg = Spec2label(
    n_encoder_inputs=8+8+3, n_outputs=2, 
    channels=128, n_heads=8, n_layers=8,
).to(device)

model_mohaom = Spec2label(
    n_encoder_inputs=8+8+3+2, n_outputs=2, 
    channels=128, n_heads=8, n_layers=8,
).to(device)

model_dir  = "/data/jdli/gaia/model/1116/"
model_name = ["sp2tefflogg_mse_A_ep500.pt", 
              "sp2mohaom_mse_A_ep500.pt"]

model_tefflogg.load_state_dict(torch.load(model_dir+model_name[0]))
model_mohaom.load_state_dict(torch.load(model_dir+model_name[1]))
```

```python
from tqdm import tqdm

def predict_4d(data_loader, model1, model2):
    
    out_lst, y_lst = np.array([]), np.array([])
    id_lst = []
    attn_lst1, attn_lst2 = np.array([]), np.array([])

    with torch.no_grad():
        for data in tqdm(data_loader):

            output1 = model1(data['x'])
            output2 = model2(torch.concat((data['x'], output1.view(-1, 1, 2)), dim=2),)
            out_lst = np.append(out_lst, torch.concat((output1, output2), dim=1).cpu().numpy())
            y_lst   = np.append(y_lst, data['y'].cpu().numpy())
            
            # attn_lst1 = np.append(attn_lst1, attn1.cpu().numpy())
            # attn_lst2 = np.append(attn_lst2, attn2.cpu().numpy())
            
            id_batch =  list(np.int64(data['id']))
            
            del output1, output2, data
            
            for idl in id_batch:
                id_lst.append(idl)

    out_lst, y_lst = np.array(out_lst).reshape(-1,4), np.array(y_lst).reshape(-1,4)
    # attn_lst1 = attn_lst1.reshape(-1, )
    # return out_lst, y_lst, id_lst
    return {'pred':out_lst, 'true':y_lst, 'id':id_lst}


data_infer = predict_4d(gdata_loader, model_tefflogg, model_mohaom)
```

```python
teff_max = gdata.df['TEFF'].max()

df_result = pd.DataFrame(
    {'source_id':np.array([np.int64(i) for i in data_infer['id']]),
    'teff_trans':teff_max*np.array(data_infer['pred'][:,0]),
    'logg_trans':np.array(data_infer['pred'][:,1]),
     'moh_trans':np.array(data_infer['pred'][:,2]),
     'aom_trans':np.array(data_infer['pred'][:,3]),
    }
)
df_result["source_id"] = pd.to_numeric(df_result["source_id"])
print(df_result.shape)

df_result = df_result.merge(df_raw, left_on="source_id", right_on="GAIAEDR3_SOURCE_ID")
print(df_result.shape, df_result.columns)
```

```python
from mpl_toolkits.axes_grid1 import make_axes_locatable
from sklearn.metrics import mean_squared_error, mean_absolute_error

cmap = plt.get_cmap('cmr.dusk')

def draw_compare(ax, pred, true, xrange=[-2, 0.5], bins=100):
    
    xx = np.linspace(xrange[0], xrange[1])
    
    res = pred-true
    
    ax.plot(xx, xx, ls='--', lw=5, c='k', zorder=5)   
    ax.hist2d(true,pred, bins=bins, cmap=cmap, zorder=4, norm=LogNorm())
    ax.set_xlim(xrange);
    ax.set_ylim(xrange);
    
    ax.text(0.6, 0.2, "RMSE = %.2f"%(mean_squared_error(true, pred, squared=False)),
            transform=ax.transAxes, zorder=3)
    ax.text(0.6, 0.1, " MAE  = %.2f"%(mean_absolute_error(true, pred)),
            transform=ax.transAxes, zorder=3)

    divider = make_axes_locatable(ax)
    ax2 = divider.append_axes("bottom", size="33%", pad=0)

    ax.figure.add_axes(ax2)
    ax2.hist2d(true, res, cmap=cmap, bins=bins, zorder=5, norm=LogNorm())
    ax2.axhline(y=0, c='k', zorder=6, lw=5, ls="--")
    ax2.set_xlabel(r"APOGEE");
    ax2.set_ylim(xrange);
    ax2.set_xlim(xrange);
    ax.set_xticks([]);
    
    return ax, ax2

def draw_res(ax, true, res, xrange=[-2, 0.5], yrange=[-0.5, 0.5]):
    ax.hist2d(true, res, bins=100, cmap=cmap, zorder=4, norm=LogNorm())
    ax.set_xlim(xrange);
    ax.set_ylim(yrange);
    ax.axhline(y=0, c='k', zorder=6, lw=5, ls="--")
    return ax
```

```python
fig, axs = plt.subplots(1,4, figsize=(5*4, 5*1))

pred_names = ['teff_trans', 'logg_trans', 'moh_trans', 'aom_trans']
true_names = ['TEFF', 'LOGG', 'M_H', 'ALPHA_M']
xranges = [[3300, 6500], [-0.1, 5], [-2.2, 0.5], [-0.3, 0.5]]
yranges = [[-1000, 1000],  [-1, 1],   [-0.7, 0.7], [-0.5, 0.5]]
xlabels = ['Teff: APO [K]', 'Log g: APO', '[M/H]: APO', r"[$\alpha$/M] : APO"]
ylabels = ['Teff: GXP [K]', 'Log g: GXP', '[M/H]: GXP', r"[$\alpha$/M] : GXP"]
y2labels = [r'$\Delta$ Teff', r'$\Delta$ Log g', r'$\Delta$ [M/H]', r'$\Delta$ [$\alpha$/M]']

for i in range(len(axs)):
    ax1, ax2 = draw_compare(
        axs[i], df_result[pred_names[i]], df_result[true_names[i]],
        xrange=xranges[i]
    )
    ax1.set_ylabel(ylabels[i]);
    ax1.set_xticks([]);
    
    ax2.set_xlabel(xlabels[i]);
    ax2.set_ylabel(y2labels[i]);
    ax2.set_ylim(yranges[i])


plt.tight_layout()
```

```python
def draw_moh_aom(ax, moh, aom, bins=100):
    
    _, _, _, img = ax.hist2d(
        moh, aom,
        bins=bins, norm=LogNorm(), 
        cmap=cmap
    )
    clb1 = fig.colorbar(img, ax=ax,);
    clb1.set_label('Counts', labelpad=-20, y=1.1, rotation=0)

    ax.set_xlabel("[M/H] : GXP");
    ax.set_ylabel(r"[$\alpha$/M] : GXP");
    ax.set_xlim([-2.5, 0.5]);
    ax.set_ylim([-0.3, 0.5]);
    return ax


def plot_kiel(ax, teff, logg, moh, bins=100):
    
    teff_bins = np.linspace(3300, 6500, 50)
    logg_bins = np.linspace(0, 5, 50)

    img = ax.scatter(teff, logg, cmap='coolwarm', s=3, c=moh, vmin=-1, vmax=0, alpha=0.5)
    
    ax.set_xlim([6500, 3300]);
    ax.set_ylim([5, -0.2]);

    clb = fig.colorbar(img, ax=ax, extend='both');
    clb.set_label(r'[M/H]$_{\rm GXP}$', labelpad=-20, y=1.1, rotation=0, )
    ax.set_xlabel("Teff : GXP");
    ax.set_ylabel("Log g: GXP");
    return ax
```
